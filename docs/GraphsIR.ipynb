```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Data provided by the user
data = {
    "Method": ["Cos Sim", "Cos Sim + Stem", "TF-IDF", "BM25", "BM25 + Stem", "BM25 + Optimizations"],
    "RQ": [0.232, 0.237, 0.197, 0.253, 0.261,0.34]
}

df = pd.DataFrame(data)

# Plotting with seaborn
plt.figure(figsize=(10, 6))
ax = sns.barplot(x="Method", y="RQ", data=df, palette="rocket", hue="Method",legend=False)
plt.xlabel('Method')
plt.ylabel('Results Quality (RQ)')
plt.xticks(rotation=0, ha="center")
plt.title('Engine Performance for each major version')

# Adding the value on top of each bar
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.3f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 9),
                textcoords = 'offset points')

plt.tight_layout()  # Adjust layout to make room for the rotated x-axis labels
plt.show()
```
